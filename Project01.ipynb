```python
hub_module = "https://tfhub.dev/deepmind/mil-nce/s3d/1"
hub_model = hub.load(hub_module)

def generate_video_embeddings(model, input_frames):
  vision_output = model.signatures["video"](tf.constant(tf.cast(input_frames, dtype = tf.float32)))
  return vision_output["video_embedding"]
  
def generate_text_embedding(model, input_words):
  text_output = model.signatures["text"](tf.constant(input_words))
  return text_output["text_embedding"]
```

```python
def crop_center_square(frame):
  y,x = frame.shape[0:2]
  min_dim = min(y, x)
  start_x = (x // 2) - (min_dim // 2)
  start_y = (y // 2) - (min_dim // 2)
  return frame[start_y:start_y + min_dim, start_x:start_x + min_dim]

def load_video(video_url, max_frames = 64, resize=(224, 224)):
  # path = tf.keras.utils.get_file(os.path.basename(video_url)[-128:], video_url)
  cap = cv2.VideoCapture("/content/build/videos/" + video_url)
  frames = []

  try:
    while True:
      ret, frame = cap.read()
      
      if not ret:
        break
      frame = crop_center_square(frame)
      frame = cv2.resize(frame, resize)
      frame = frame[:, :, [2, 1, 0]]
      frames.append(frame)

      if len(frames) == max_frames:
        break;

  finally:
    cap.release()

  frames = np.array(frames)

  if len(frames) < max_frames:
    n_repeat = int(math.ceil(max_frames/float(len(frames))))
    frames = frames.repeat(n_repeat, axis = 0)

  frames = frames[:max_frames]
  return frames / 255.0

video_1_url = 'football.mp4'
video_2_url = 'song.mp4' 
video_3_url = "reading.mp4"
video_4_url = "food.mp4"
video_5_url = "guiter.mp4"
video_6_url = "cricket.mp4"

video_1 = load_video(video_1_url)
video_2 = load_video(video_2_url)
video_3 = load_video(video_3_url)
video_4 = load_video(video_4_url)
video_5 = load_video(video_5_url)
video_6 = load_video(video_6_url)

all_videos_urls = [video_1_url, video_2_url, video_3_url, 
                   video_4_url, video_5_url, video_6_url]
all_videos = [video_1, video_2, video_3, video_4, video_5, video_6]
```

```python
videos_np = np.stack(all_videos, axis = 0)

video_embd = generate_video_embeddings(hub_model, videos_np)
```

```python
server = None

server = Flask(__name__, template_folder='/content/build')
server.config['CORS_HEADERS'] = 'Content-Type'
cors = CORS(server)

# run_with_ngrok(server)   
```

```python
@server.route('/static/<folder_name>/<file_name>')
def send_js(folder_name, file_name):
  
    return send_from_directory('/content/build/static/' + folder_name, file_name)

@server.route('/videos/<file_name>')
def get_video(file_name):
  
    return send_from_directory('/content/build/videos', file_name)
```

```python
@server.route("/")
@cross_origin()
def re():
    return render_template('index.html')
```

```python
@server.route("/qa", methods=['POST'])
@cross_origin()
def hello():
  data = request.get_json()
  
  question = data['question']
  paragraph = data['paragraph']
  
  if question.strip() == "" or paragraph.strip() == "":
      return jsonify({ "error": "Question and paragraph can't be null" })

  question_tokens = tokenizer.tokenize(question)
  paragraph_tokens = tokenizer.tokenize(paragraph)
  tokens = ['[CLS]'] + question_tokens + ['[SEP]'] + paragraph_tokens + ['[SEP]']
  input_word_ids = tokenizer.convert_tokens_to_ids(tokens)
  input_mask = [1] * len(input_word_ids)
  input_type_ids = [0] * (1 + len(question_tokens) + 1) + [1] * (len(paragraph_tokens) + 1)

  input_word_ids, input_mask, input_type_ids = map(lambda t: tf.expand_dims(
      tf.convert_to_tensor(t, dtype=tf.int32), 0), (input_word_ids, input_mask, input_type_ids))
  
  outputs = model([input_word_ids, input_mask, input_type_ids])

  # using `[1:]` will enforce an answer. `outputs[0][0][0]` is the ignored '[CLS]' token logit
  short_start = tf.argmax(outputs[0][0][1:]) + 1
  short_end = tf.argmax(outputs[1][0][1:]) + 1
  answer_tokens = tokens[short_start: short_end + 1]
  answer = tokenizer.convert_tokens_to_string(answer_tokens)

  response = jsonify({ "answer": answer})
  return response
```

```python
@server.route("/chat", methods=['POST'])
@cross_origin()
def chat():
    global chat_history_ids

    data = request.get_json()
    
    msg = data['msg']

    chat_history_ids.append(msg)

    inputs = tokenizer_chat(["</s> <s>".join(chat_history_ids)], return_tensors='tf')
    reply_ids = model_chat.generate(**inputs)
    reply = tokenizer_chat.batch_decode(reply_ids, skip_special_tokens=True)[0]
    chat_history_ids.append(reply)

    return jsonify({ "msg": reply})
```

```python
@server.route("/searchVideo", methods=['POST'])
@cross_origin()
def getTextToVideo():
  data = request.get_json()
  
  searchText = data['searchText']

  words_np = np.array([searchText])

  text_embd = generate_text_embedding(hub_model, words_np)

  all_scores = np.round(tf.math.softmax(text_embd @ tf.transpose(video_embd)).numpy(), decimals = 4).astype("str")

  result = { "videos": dict(zip(all_videos_urls, all_scores[0].tolist())) }
  return jsonify(result)
```
